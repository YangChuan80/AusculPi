```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = Image.open('3M.bmp')
image
```

```python
x = image.size[0]
y = image.size[1]
```

```python
print(x)
print(y)
```

```python
matrix = []
points = []
integrated_density = 0

for i in range(x):
    matrix.append([])
    for j in range(y):
        matrix[i].append(image.getpixel((i,j)))
        #integrated_density += image.getpixel((i,j))[1]
        #points.append(image.getpixel((i,j))[1])
```

### Extract Red Line Position

```python
redMax = 0
xStore = 0
yStore = 0
for xAxis in range(x):
    for yAxis in range(y):
        currentPoint = matrix[xAxis][yAxis]
        if currentPoint[0] >= redMax and currentPoint[1] < 10 and currentPoint[2] < 10:
            redMax = currentPoint[0]
            xStore = xAxis
            yStore = yAxis
            
print(xStore, yStore)
```

- The redline position is located at y = 252.

```python
absMax = 0
arr = []
points_vertical = []
theOne = 0

for xAxis in range(x):
    for yAxis in range(y):
        currentPoint = matrix[xAxis][yAxis]
        # Pickup Blue points
        if currentPoint[2] == 255 and currentPoint[0] < 200 and currentPoint[1] < 200:
            points_vertical.append(yAxis)
            
    #print(points_vertical)
    
    
    # Choose the largest amplitude
    for item in points_vertical:
        
        if abs(item-252) > absMax:
            absMax = abs(item-252)
            theOne = item 
    arr.append(theOne)
    
    absMax = 0  
    theOne = 0
    points_vertical = []            
arr
```

```python
fig = plt.figure()
s = fig.add_subplot(111)
s.plot(arr, linewidth=0.6, color='blue')
```
